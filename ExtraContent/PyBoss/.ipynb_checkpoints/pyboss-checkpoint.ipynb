```python
import pandas as pd
import us_state_abbrev as us
```

```python
file = 'employee_data.csv'

employee_df = pd.read_csv(file)
employee_df.head()
```

```python
cols = list(employee_df.columns.values)
cols
```

```python
new_cols = employee_df['Name'].str.split(' ', n = 1, expand = True)
employee_df['First Name'] = new_cols[0]
employee_df['Last Name'] = new_cols[1]
employee_df.head()
```

```python
columns_formatted = ['Emp ID', 'First Name', 'Last Name', 'DOB', 'SSN', 'State']

employee_formatted_df = employee_df.reindex(columns=columns_formatted)
employee_formatted_df.head()
```

```python
checker = pd.to_datetime(employee_formatted_df['DOB'])
checker = checker.dt.strftime('%m/%d/%Y')
checker.head()

employee_formatted_df['DOB'] = checker
employee_formatted_df['DOB'].head()
```

```python
employee_formatted_df.head()
```

```python
# WHY CANT THIS WORK WAAAAAAAAHHH
#(pd.Series(employee_formatted_df['SSN']).str.replace(all_ssn[i],'***-**',n=1, regex = False))

# LIST! COMPREHENSION! 4! DA! WIN!!
fixed_ssn = [(employee_formatted_df['SSN'][i].replace(employee_formatted_df['SSN'][i][:6], '***-**')) for i in range(len(employee_formatted_df['SSN']))]
employee_formatted_df['SSN'] = fixed_ssn
employee_formatted_df.head()
```

```python
fixed_state = [(us.us_state_abbrev[x]) for x in employee_formatted_df['State']]
employee_formatted_df['State'] = fixed_state

employee_formatted_df.head()
```

```python
employee_formatted_df.to_csv('employee_formatted.csv', index = False, header=True)
```
